```python
#import libraries

import pandas as pd
import numpy as np
import tweepy
import requests
import json
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline
```

```python
#load the final dataset
master_data = pd.read_csv('twitter_archive_master.csv')
```

## Insight 1

### 1.Which breed has maximum count in prediction 1 column?

```python
#calculate the counts of dog's type in predection 1 by using value count function
master_data.prediction_1.value_counts()
```

### So from output it's clear that golden retriever has maximum number of prediction 1 count.    

## Insight 2

### 2. what's the correlation bettwen number of retweets and number of favorite ?

```python
#Plot scatter plot between favorite count and retweet count
plt.scatter(data=master_data,x='retweet_count',y='favorite_count',alpha=0.7)
plt.xlabel('number of favorites')
plt.ylabel('number of retweets')
plt.title("Scatter plot between favorite count and retweet count")
```

### There is a positive correlation between favorite count and retweet count,thus with  increase in the number of favorite, we can see that number of retweet  also increases.

## Insight 3

###  3. Which stage of dog got more number of tweets?

```python
#plot bar graph of Dog's stage
master_data.dog_stage.value_counts().plot(kind='bar')
plt.xlabel('Dog stage')
plt.ylabel('count')
plt.title("Bar Graph of different stage")
```

### Pupper stage dogs got more than 200 tweets.Hence pupper got most of the tweets.

## Insight 4

### 4. Which breed of dog got Maximum number of retweets and favorites count?

```python
#Sort retweet count in descending order 
master_data.sort_values(by='retweet_count',ascending=False)['prediction_1']
```

#### From output it's clear that Labrador retriever breed of dog  got most of the retweet.

```python
master_data.sort_values(by='favorite_count',ascending=False)['prediction_1']
```

#### From output it's clear that Labrador retriever breed of dog  got most of the favorite count.


###  Thus  Labrador_retriever breed has maximum no of favorite count and retweet count on Twitter
